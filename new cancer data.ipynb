```python
import pandas as pd

# filenames
excel_names = ["1.xlsx", "2.xlsx", "3.xlsx","4.xlsx", "5.xlsx", "6.xlsx","7.xlsx", "8.xlsx", "9.xlsx","10.xlsx", "11.xlsx", "12.xlsx","13.xlsx", "14.xlsx", "15.xlsx","16.xlsx", "17.xlsx", "18.xlsx","19.xlsx", "20.xlsx", "21.xlsx","22.xlsx", "23.xlsx", "24.xlsx","25.xlsx"]

# read them in
excels = [pd.ExcelFile(name) for name in excel_names]

# turn them into dataframes
frames = [x.parse(x.sheet_names[0], header=None,index_col=None) for x in excels]

# delete the first row for all frames except the first
# i.e. remove the header row -- assumes it's the first
frames[1:] = [df[1:] for df in frames[1:]]

# concatenate them..
combined = pd.concat(frames)

# write it out
combined.to_excel("c.xlsx", header=False, index=False)
```
